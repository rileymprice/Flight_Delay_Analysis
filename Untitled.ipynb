```python
import pandas as pd
from pathlib import Path
import feather
%matplotlib inline
```

```python
#all_flight_data = pd.read_feather(Path('Data/Processed/all_flight_data.feather'),nthreads=1)
all_flight_data = feather.read_dataframe(Path('Data/Processed/all_delayed_flight_data.feather'))
```

```python
all_flight_data = all_flight_data.set_index('Airline')
all_flight_data = all_flight_data.drop(columns=['index'])
all_flight_data.head()
```

```python
delayed_flight_data = all_flight_data[all_flight_data['DepDelay'] > 0]
```

```python
delays_per_origin = all_flight_data.groupby('Origin')['DepDelay'].sum().sort_index(ascending=False)
delays_per_origin = delays_per_origin.to_frame()
tmp_flight_count= all_flight_data['Origin'].value_counts().sort_index(ascending=False)
delays_per_origin['FlightCount'] = tmp_flight_count.values
delays_per_origin['FlightCount'].describe()
```

```python
delays_per_origin['AverageDelay'] = delays_per_origin['DepDelay'] / delays_per_origin['FlightCount']
delays_per_origin = delays_per_origin.sort_values(by=['AverageDelay'],ascending=False)
delays_per_origin[delays_per_origin['FlightCount'] > 1000]['AverageDelay'].plot.bar(title='Average delay per flight per origin')
```

```python
number_of_delayed_flights_per_origin = delayed_flight_data['Origin'].value_counts().sort_index(ascending=False).to_frame()
number_of_delayed_flights_per_origin = number_of_delayed_flights_per_origin.reset_index()
number_of_delayed_flights_per_origin.columns = ['Origin','DelayedFlightCount']
number_of_delayed_flights_per_origin.set_index('Origin',inplace=True)
number_of_delayed_flights_per_origin['TotalFlightCount'] = all_flight_data['Origin'].value_counts().sort_index(ascending=False).values
number_of_delayed_flights_per_origin
```
